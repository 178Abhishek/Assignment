```python
'''Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.
AnS:-Matplotlib is a Python library used for creating static, interactive, and animated visualizations in Python. It provides a wide range of functionalities for generating plots, histograms, bar charts, scatter plots, and many other types of visualizations.

It is used for data visualization tasks, allowing users to represent data in a graphical format, which aids in data analysis, understanding patterns, and communicating insights effectively.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line plot: Displaying data points as markers connected by straight lines.
Histogram: Showing the distribution of numerical data by dividing the data into bins and displaying bars of frequency.
Bar plot: Representing categorical data with rectangular bars, where the height or length of the bar corresponds to the value of the data.
Scatter plot: Displaying data points as individual markers, where the position of each marker corresponds to the value of two variables.
Pie chart: Illustrating numerical proportions as slices of a circular pie, where each slice represents a category and its size corresponds to 
the proportion of the data it represents.'''
```

```python
'''Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

ANS:- A scatter plot is a type of data visualization that displays the values of two variables as points on a two-dimensional coordinate system.
Each point on the plot represents an observation in the dataset, with the x-coordinate corresponding to the value of one variable 
and the y-coordinate corresponding to the value of the other variable.
Scatter plots are useful for visually identifying relationships or patterns between the variables, 
such as correlations or trends'''

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

#Note: Also add title, xlabel, and ylabel to the plot.

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('This Title of My Graph')
plt.show()
```

```python
'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])'''

'''Ans:- 
A bar plot is a type of data visualization that represents categorical data with rectangular bars. The length or height of each bar corresponds to the value of the data it represents. Bar plots are commonly used to compare the values of different categories, track changes over time, or illustrate distributions.

Bar plots are used for various purposes, including:

Comparison: Bar plots make it easy to compare the values of different categories visually. By representing data as bars, it becomes straightforward to see which category has higher or lower values.

Trend Analysis: Bar plots can be used to track changes in values over time or across different groups. Trends and patterns in the data can be easily identified through bar plots.

Presentation: Bar plots are effective tools for presenting data in reports, presentations, and academic papers'''

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.title('Profit of Companies')
plt.grid(True)
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='salmon')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.title('Profit of Companies')
plt.grid(True)
plt.show()
```

```python
'''Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)'''

'''Ans:- A box plot, also known as a box-and-whisker plot, is a statistical visualization method used to display the distribution of a dataset. It provides a summary of the dataset's central tendency, dispersion, and skewness. Box plots are particularly useful for identifying outliers and comparing the distributions of different datasets.

Here's why box plots are used:

Summary of Distribution: Box plots provide a concise summary of the distribution of a dataset, including the median, quartiles, and potential outliers.

Comparison of Distributions: Box plots enable easy comparison of the distributions of multiple datasets, allowing users to identify differences in central tendency, spread, and variability.

Identification of Outliers: Outliers, or data points that deviate significantly from the rest of the dataset, are visually identifiable in box plots, making them useful for outlier detection.

Robustness: Box plots are robust to variations in the shape of the distribution and are not affected by extreme values as much as other visualization methods'''

import numpy as np
import matplotlib.pyplot as plt

# Generate random data for box1 and box2
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a single list
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Data')
plt.xlabel('Box')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
