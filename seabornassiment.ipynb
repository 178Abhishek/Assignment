```python
'''Que 1: Name any five plots that we can plot using the Seaborn library.
Also, state the uses of each plot.'''

'''Ans:-Seaborn is a powerful Python visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive statistical graphics. Here are five plots that can be plotted using Seaborn along with their uses:

Histograms:
Use: Histograms are used to visualize the distribution of a univariate dataset. They display the frequency distribution of continuous data by dividing the data into bins and showing the number of observations in each bin.
Bar Plots:
Use: Bar plots are used to represent categorical data with rectangular bars. They are effective for comparing the values of different categories and visualizing the distribution or composition of a dataset.
Box Plots (Box-and-Whisker Plots):
Use: Box plots are used to visualize the distribution of a dataset and display its five-number summary: minimum, first quartile, median, third quartile, and maximum. They are useful for identifying outliers and comparing the spread and central tendency of multiple datasets.
Violin Plots:
Use: Violin plots are similar to box plots but provide a richer visualization of the distribution of the data. They combine a box plot with a kernel density estimation (KDE) plot, allowing for a more comprehensive understanding of the data distribution, including its shape and multimodality.
Scatter Plots:
Use: Scatter plots are used to visualize the relationship between two continuous variables. They display individual data points as markers on a two-dimensional plane, making it easy to identify patterns, trends, correlations, and outliers in the data.'''
```

```python
'''Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.'''

'''Ans:-'''
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot a line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)
plt.title('FMRI Signal over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event')
plt.show()
```

```python
'''Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.'''
'''Ans:-'''

import seaborn as sns
import matplotlib.pyplot as plt
# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Box plot for age vs. pclass
sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axes[0])
axes[0].set_title('Age vs. Pclass')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Age')

# Box plot for fare vs. pclass
sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axes[1])
axes[1].set_title('Fare vs. Pclass')
axes[1].set_xlabel('Pclass')
axes[1].set_ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
'''Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

Ans:- '''
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue for the 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x="price", hue="cut", multiple="stack", bins=30, kde=False, palette='husl')
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.show()


```

```python
'''Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

Ans:- '''
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for the 'species' column
sns.pairplot(iris_data, hue="species", palette="husl")
plt.show()

```

```python
'''Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

Ans:- '''
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Number of Passengers by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
